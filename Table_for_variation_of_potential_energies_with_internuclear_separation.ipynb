```python
#HCl Table Generating Code
# Table generating code for HCl
import numpy as np
import pandas as pd

Re_g = 1.274
Re_e = 1.55
k_g = 5.1
k_e = 4.2
shift = 8.6

R = np.linspace(0.8, 2.5, 30)

V_ground = 0.5 * k_g * (R - Re_g)**2
V_excited = 0.5 * k_e * (R - Re_e)**2 + shift

df = pd.DataFrame({
    "R (Å)": np.round(R, 3),
    "V_ground (eV)": np.round(V_ground, 4),
    "V_excited (eV)": np.round(V_excited, 4)
})

print("="*40)
print("Parabolic Table for HCl")
print(df.to_string(index=False))

df.to_csv("HCl_Harmonic_Table.csv", index=False)

```
